```python
import matplotlib.pyplot as plt
```

```python
path = "/home/sante/Documents/Laboratorio CPD/for-loop/first.txt"
path2 = "/home/sante/Documents/Laboratorio CPD/for-loop/second.txt"

x = []
y = []
for line in open(path, 'r'):
    lines = [i for i in line.split()]
    x.append(lines[0])
    y.append(int(lines[1]))

a = []
b = []
for line in open(path2, 'r'):
    lines = [i for i in line.split()]
    a.append(lines[0])
    b.append(int(lines[1]))
      
plt.title("")
plt.xlabel('Tamanho')
plt.ylabel('Time')
plt.yticks(y)
plt.yticks(b)
plt.plot(x, y, marker = 'o', c = 'g', label="map-row")
plt.plot(a, b, marker = 'o', c = 'r', label="map-column")
plt.legend()
plt.show()
```
