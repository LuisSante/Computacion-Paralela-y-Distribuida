```python
import matplotlib.pyplot as plt
```

```python
path = "/home/sante/Documents/Laboratorio CPD/matrix/classic.txt"
path2 = "/home/sante/Documents/Laboratorio CPD/matrix/block.txt"

x = []
y = []
for line in open(path, 'r'):
    lines = [i for i in line.split()]
    x.append(lines[0])
    z = float(lines[1])
    y.append(int(z))

a = []
b = []
for line in open(path2, 'r'):
    lines = [i for i in line.split()]
    a.append(lines[0])
    c = float(lines[1])
    b.append(int(c))
      
plt.title("")
plt.xlabel('Tamanho')
plt.ylabel('Time')
plt.yticks(y)
plt.yticks(b)
plt.plot(x, y, marker = 'o', c = 'g', label="classic")
plt.plot(a, b, marker = 'o', c = 'r', label="block")
plt.legend()
plt.show()
```
